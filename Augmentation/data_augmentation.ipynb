```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE
```

```python
df=pd.read_csv("../DataWork/transformed_df.csv")
df
```

```python
from sklearn.impute import KNNImputer

def fill_na_with_knn(df):
    # Separate the DataFrame into numeric and categorical columns
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = df.select_dtypes(include=['object', 'bool']).columns

    # Fill numeric NaN values with KNN imputation
    imputer_numeric = KNNImputer(n_neighbors=5)
    df_numeric = df[numeric_cols].copy()
    df[numeric_cols] = imputer_numeric.fit_transform(df_numeric)

    # Fill categorical NaN values with mode imputation
    for col in categorical_cols:
        df[col].fillna(df[col].mode()[0], inplace=True)

    return df

df = fill_na_with_knn(df)
```

```python
# Separate the features and target variable
X = df.drop(['diagnostic',"img_id", "image_path"], axis=1)
y = df['diagnostic']
```

```python

# Calculate the current class distribution
class_distribution = y.value_counts()
current_mel_count = class_distribution['MEL']

# Calculate the desired count of 'MEL' class after augmentation
desired_mel_count = current_mel_count + 108

# Create a dictionary with the desired sampling strategy for each class
sampling_strategy = {label: desired_mel_count if label == 'MEL' else count for label, count in class_distribution.items()}

# Apply SMOTE to generate additional rows for 'MEL' class
smote = SMOTE(sampling_strategy=sampling_strategy)
augmented_X, augmented_y = smote.fit_resample(X, y)

# Create a new DataFrame with the augmented data
augmented_df = pd.DataFrame(augmented_X, columns=X.columns)
augmented_df['diagnostic'] = augmented_y

# Verify the class distribution
class_distribution = augmented_df['diagnostic'].value_counts()
print(class_distribution)
```

```python
import pandas as pd

# Assuming your augmented dataframe is named 'augmented_df'

# Create an empty dataframe for the training set
train_df = pd.DataFrame()

# Create an empty dataframe for the test set
test_df = pd.DataFrame()

# Iterate over each class in the augmented dataframe
for class_label in augmented_df['diagnostic'].unique():
    # Filter the rows for the current class
    class_rows = augmented_df[augmented_df['diagnostic'] == class_label]
    
    # Randomly select 150 rows for the training set
    train_rows = class_rows.sample(n=150, random_state=42)
    
    # Add the selected rows to the training set
    train_df = pd.concat([train_df, train_rows])
    
    # Remove the selected rows from the original dataframe
    class_rows = class_rows.drop(train_rows.index)
    
    # Randomly select 10 rows for the test set
    test_rows = class_rows.sample(n=10, random_state=42)
    
    # Add the selected rows to the test set
    test_df = pd.concat([test_df, test_rows])


#train_df.to_csv("train_df_2.csv", index=False)
#test_df.to_csv("test_df_2.csv", index=False)
```
